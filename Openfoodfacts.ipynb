```python
food.drop(columns=['carbon-footprint_100g','glycemic-index_100g','water-hardness_100g'],inplace=True,axis =1 )
```

```python
continuous_cols = food.select_dtypes(include='number').columns
correlation_matrix = food[continuous_cols].corr()
plt.figure(figsize=(12, 10)) # size
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```
